```python
class Solution:
    # Function to sort an array of 0s, 1s, and 2s
    def sort012(self, arr):
        low = 0
        mid = 0
        high = len(arr) - 1

        while mid <= high:
            if arr[mid] == 0:
                arr[low], arr[mid] = arr[mid], arr[low]
                low += 1
                mid += 1
            elif arr[mid] == 1:
                mid += 1
            else:  # arr[mid] == 2
                arr[mid], arr[high] = arr[high], arr[mid]
                high -= 1

        return arr  # Return the sorted array
```

```python
class Solution:
    def binarysearch(self, arr, k):
        low, high = 0, len(arr) - 1
        result = -1
        
        while low <= high:
            mid = (low + high) // 2
            
            if arr[mid] == k:
                result = mid      # found k, record index
                high = mid - 1    # continue search on left side for first occurrence
            elif arr[mid] < k:
                low = mid + 1
            else:
                high = mid - 1
        
        return result
```
