```python
# 9.1 Define the good() function
def good():
    return ['Harry', 'Ron', 'Hermione']

# 9.2 Define the get_odds() generator
def get_odds():
    for number in range(10):
        if number % 2 != 0:
            yield number

# Run and print results
print("good() returns:", good())

# Find and print the third odd number
count = 0
for odd in get_odds():
    count += 1
    if count == 3:
        print("Third odd number is:", odd)
        break
```
