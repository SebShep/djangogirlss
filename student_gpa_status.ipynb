```python
# Name: Sebastian Shepard
# File Name: student_gpa_checker
# Description: This app accepts student names and GPAs, then evaluates whether each student qualifies
# for the Dean's List (GPA ≥ 3.5) or the Honor Roll (GPA ≥ 3.25). The input ends when 'ZZZ' is entered as a last name.

def check_student_status():
    while True:
        last_name = input("Enter the student's last name (or 'ZZZ' to quit): ")
        if last_name.upper() == 'ZZZ':
            print("Finished processing student records.")
            break

        first_name = input("Enter the student's first name: ")
        try:
            gpa = float(input("Enter the student's GPA: "))
        except ValueError:
            print("Invalid GPA entered. Please use a decimal number.")
            continue

        if gpa >= 3.5:
            print(f"{first_name} {last_name} has made the Dean's List.\n")
        elif gpa >= 3.25:
            print(f"{first_name} {last_name} has made the Honor Roll.\n")
        else:
            print(f"{first_name} {last_name} did not qualify for honors.\n")

# Run the app
if __name__ == "__main__":
    check_student_status()
```
