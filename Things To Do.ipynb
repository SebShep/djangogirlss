```python
# 4.1 - Number Guessing Game
secret = 7
guess = 5

if guess < secret:
    print("too low")
elif guess > secret:
    print("too high")
else:
    print("just right")
```

```python
# 4.2 - Match Fruit Based on small and green
small = True
green = True

if small and green:
    print("pea")
elif small and not green:
    print("cherry")
elif not small and green:
    print("watermelon")
else:
    print("pumpkin")
```

```python
# 6.1 - For loop to print values from the list
for value in [3, 2, 1, 0]:
    print(value)
```

```python
# 6.2 - While loop to compare number with guess_me
guess_me = 7
number = 1

while True:
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    else:
        print("oops")
        break
    number += 1
```

```python
# 6.3 - For loop with conditions using range
guess_me = 4

for number in range(10):
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    else:
        print("oops")
        break
```
